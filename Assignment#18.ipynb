# Forecast on Coca-Cola Prices

```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_excel('CocaCola_Sales_Rawdata.xlsx')
df.head()
```

```python
df.info()
```

```python
df.index=df.Quarter
```

```python
df.drop('Quarter',axis=1)
```

```python
df1=df.copy()
```

```python
df1['range']=pd.date_range(start='1/1/1986', periods=len(df1),freq='Q')
```

```python
df1
```

```python
df1.drop('Quarter',axis=1)
```

```python
df1['day']=df1.range.dt.day
```

```python
df1['month']=df1.range.dt.month
df1['year']=df1.range.dt.year
```

```python
df1.drop('Quarter',axis=1)
```

```python
df2=pd.get_dummies(df1['month'])
```

```python
df2['t']=np.arange(1,43)
```

```python
df2.shape
```

```python
df2
```

```python
df3=pd.concat([df1['Sales'],df2],axis=1)
```

```python
df3
```

```python
df3['t_square']=df3['t']*df3['t']
```

```python
df3['Sales']=np.log(df3['Sales'])
```

```python
df3
```

```python
df3['log_s']=np.log(df3['Sales'])
```

```python
df3
```

```python
df3.columns=['Sales','Mar','Jun','Sep','Dec','t','t_square','log_s']
```

```python
df3
```

```python
df3.shape
```

```python
Train=df3.head(30)
Train
```

```python
Test=df3.tail()
```

```python
import statsmodels.formula.api as smf 
```

## Linear Model 

```python
linear_model = smf.ols('Sales~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

## Exponential Model

```python
Exp = smf.ols('log_s~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

## Quadratic Model

```python
Quad = smf.ols('Sales~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

## Additive Seasonality

```python
add_sea = smf.ols('Sales~Mar+Jun+Sep',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Mar','Jun','Sep']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

## Additive Seasonality Quadratic

```python
add_sea_Quad = smf.ols('Sales~t+t_square+Mar+Jun+Sep',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['Mar','Jun','Sep','t','t_square']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

## Multiplicative Seasonality

```python
Mul_sea = smf.ols('log_s~Mar+Jun+Sep',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

## Multiplicative Additive Seasonality

```python
Mul_Add_sea = smf.ols('log_s~t+Mar+Jun+Sep',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea 
```

```python
models= {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
```

```python
pd.DataFrame(models)
```

As the RMSE value for the Multiplicative Additive Seasonality is low. It is best to use this model. 

# Forecast on Airlines Passengers

```python
data.info()
```

```python
data.index=data.Month
```

```python
data
```

```python
data1=data.drop('Month',axis=1)
```

```python
data1
```

```python
data1['whole']=data1.index
```

```python
data1
```

```python
data1['month']=data1.whole.dt.month
data1['year']=data1.whole.dt.year
data1['day']=data1.whole.dt.day
```

```python
data1
```

```python
lill=pd.get_dummies(data1.month)
lill
```

```python
data3=pd.concat([data1.Passengers,lill],axis=1)
data3
```

```python
data3['t']=np.arange(1,97)
```

```python
data3['t_square']=data3['t']*data3['t']
```

```python
data3['log_p']=np.log(data3['Passengers'])
```

```python
data3
```

```python
data3.columns=['Passengers','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','t','t_square','log_p']
```

```python
Train2=data3.head(75)
```

```python
Train2
```

```python
Test2=data3.tail(21)
```

```python
Test2
```

```python
linear_model2 = smf.ols('Passengers~t',data=Train2).fit()
```

```python
linear_model2.summary()
```

```python
predict2=linear_model2.predict(Test2['t'])
```

```python
pred2=pd.DataFrame(predict2)
```

```python
def rmse(actual,predict):
    temp=np.sqrt(np.mean((actual-predict)*(actual-predict)))
    return temp
```

```python
rmse_linear2=rmse(Test2.Passengers,predict2)
rmse_linear2
```

```python
Exp2 = smf.ols('log_p~t',data=Train2).fit()
pred_Exp2 = pd.Series(Exp2.predict(pd.DataFrame(Test2['t'])))
rmse_Exp2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(np.exp(pred_Exp2)))**2))
rmse_Exp2

```

```python
Quad2 = smf.ols('Passengers~t+t_square',data=Train2).fit()
pred_Quad2 = pd.Series(Quad2.predict(Test2[["t","t_square"]]))
rmse_Quad2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(pred_Quad2))**2))
rmse_Quad2
```

```python
add_sea2 = smf.ols('Passengers~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=Train2).fit()
pred_add_sea2 = pd.Series(add_sea2.predict(Test2[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))
rmse_add_sea2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(pred_add_sea2))**2))
rmse_add_sea2
```

```python
add_sea_Quad2 = smf.ols('Passengers~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=Train2).fit()
pred_add_sea_quad2 = pd.Series(add_sea_Quad2.predict(Test2[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','t','t_square']]))
rmse_add_sea_quad2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(pred_add_sea_quad2))**2))
rmse_add_sea_quad2
```

```python

Mul_sea2 = smf.ols('log_p~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data = Train2).fit()
pred_Mult_sea2 = pd.Series(Mul_sea2.predict(Test2))
rmse_Mult_sea2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(np.exp(pred_Mult_sea2)))**2))
rmse_Mult_sea2
```

```python
Mul_Add_sea2 = smf.ols('log_p~t+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data = Train2).fit()
pred_Mult_add_sea2 = pd.Series(Mul_Add_sea2.predict(Test2))
rmse_Mult_add_sea2 = np.sqrt(np.mean((np.array(Test2['Passengers'])-np.array(np.exp(pred_Mult_add_sea2)))**2))
rmse_Mult_add_sea2 
```

```python
models2= {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear2,rmse_Exp2,rmse_Quad2,rmse_add_sea2,rmse_add_sea_quad2,rmse_Mult_sea2,rmse_Mult_add_sea2])}
```

```python
pd.DataFrame(models2)
```

As the RMSE value for the Multiplicative Additive Seasonality is low. It is best to use this model.
